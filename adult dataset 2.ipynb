# train and test the data

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.linear_model import LinearRegression
rs = RandomState(13940)

features = x_adult.loc[:,'age':'native_country_Yugoslavia']
X = features.values
Y = x_adult['annual_income_>50K'].values

X_train,X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state=rs)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)

```

```python
linreg = LinearRegression()
linreg.fit(X_train, Y_train)
y_pred = linreg.predict(X_test)
print(y_pred)
```

```python
score_linreg = linreg.score(X_test, Y_test) 
print('the accuracy of the Linear Regression model', score_linreg)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(Y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
y_pred = knn.predict(X_test)
score_knn = knn.score(X_test,Y_test)
print('The accuracy of the KNN Model is',score_knn)
```

```python

from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()
DTC.fit(X_train, Y_train)
DTC_pred = DTC.predict(X_test)
score_DTC = DTC.score(X_test, Y_test)
print('The accuracy of the DTC Model is',score_DTC)
```

```python
mean_squared_error(Y_test, DTC_pred)
```
